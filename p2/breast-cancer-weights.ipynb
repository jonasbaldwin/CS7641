```python
max_trials = 10
rand_seeds = random.sample(range(1, 999999999), max_trials)
```

```python
def print_groups(g):
    for key, item in g:
        print(g.get_group(key), "\n\n")
```

```python
def show_aggregates(attempts):
    att_group = attempts.groupby('samp_size')
    att_agg = att_group['out_samp_acc'].agg([np.mean, np.std, np.median, np.min, np.max])
    att_agg.columns = ['OS Acc Mean', 'OS Acc Std.', 'OS Acc Median', 'OS Acc  Min', 'OS Acc  Max']
    time_agg = att_group['time'].agg([np.mean])
    time_agg.columns = ['Time Mean']
    data_agg_comb = pd.concat([att_agg, time_agg], axis=1, sort=False)
    return data_agg_comb
```

```python
def show_learning_rate(data, title, ylim=None, xlim=None):
    plt.plot(data['OS Acc Mean'], 'b', data['OS Acc Median'], 'r')
    plt.legend(['Out of Sample Mean Accuray', 'Out of Sample Median Accuracy'], loc='lower right')
    plt.xlabel('sample size')
    plt.ylabel('out of sample accuracy')
    if ylim is not None:
        plt.ylim(ylim[0],ylim[1])
    if xlim is not None:
        plt.xlim(xlim[0],xlim[1])
    plt.title(title)
    plt.show()
```

```python
def show_learning_time(data, title):
    plt.plot(data['Time Mean'], 'gray')
    plt.legend(['Time Mean'], loc='lower right')
    plt.xlabel('sample size')
    plt.ylabel('time')
    plt.title(title)
    plt.show()
```

```python
def show_charts(data, title):
    show_learning_rate(data, title)
    show_learning_time(data, title + ' Time')
```

```python
x = bc_data.loc[:, bc_data.columns != 'diagnosis']
y = np.where(bc_data['diagnosis']=='M', 1, 0)
```

```python
# preprocessing and training (modeled after https://mlrose.readthedocs.io/en/stable/source/tutorial3.html)
def train_and_test_model(seed, network_shape, algorithm, test_size):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state = seed)

    scaler = MinMaxScaler()

    x_train_scaled = scaler.fit_transform(x_train)
    x_test_scaled = scaler.transform(x_test)

    one_hot = OneHotEncoder(categories='auto')

    y_train_hot = one_hot.fit_transform(y_train.reshape(-1, 1)).todense()
    y_test_hot = one_hot.transform(y_test.reshape(-1, 1)).todense()

    bc_nn_model = mlrose.NeuralNetwork(hidden_nodes = network_shape, activation = 'relu', \
                                     algorithm = algorithm, max_iters = 1000, \
                                     bias = True, is_classifier = True, learning_rate = 0.0001, \
                                     early_stopping = True, clip_max = 5, max_attempts = 100, \
                                     random_state = seed)
    bc_nn_model.fit(x_train_scaled, y_train_hot)
    y_train_pred = bc_nn_model.predict(x_train_scaled)
    y_train_accuracy = accuracy_score(y_train_hot, y_train_pred)

    
    y_test_pred = bc_nn_model.predict(x_test_scaled)
    y_test_accuracy = accuracy_score(y_test_hot, y_test_pred)
                                      
    return [y_train_accuracy, y_test_accuracy]
```

```python
def tnt_alg(algorithm, test_size=0.3):
    attempts = []
    for t in range(len(rand_seeds)):
        seed = rand_seeds[t]
        network_shape = [9,9] # moderately good preforming model with 86.11% accuracy
        start_time = datetime.datetime.now()
        [os_y_acc, is_y_acc] = train_and_test_model(seed, network_shape, algorithm, test_size)
        total_delta = datetime.datetime.now() - start_time
        microseconds  = total_delta.seconds * 1000000 + total_delta.microseconds
        attempts.append([t, seed, microseconds, os_y_acc, is_y_acc, test_size])
    df = pd.DataFrame(attempts, columns=['trial-n', 'seed', 'time', 'out_samp_acc', 'in_samp_acc', 'samp_size'])
    
    return df
```

```python
def tnt_multisize_alg(algorithm):
    results = None
    for t in range(1,20):
        w = t / 20
        r = tnt_alg(algorithm, w)
        if results is None:
            results = tnt_alg(algorithm, w)
        else:
            results = results.append(r)
    results.reset_index()
    return results
```
